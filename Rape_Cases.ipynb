```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("20_Victims_of_rape.csv")
```

```
temp = df.iloc[:,[x for x in range(3,11)]]
df['SUM'] = temp.sum(axis=1)
```

```
onlyStates = df[~df['Area_Name']
                  .isin(['Andaman & Nicobar Islands','Chandigarh','Dadra & Nagar Haveli','Daman & Diu','Jammu & Kashmir','Lakshadweep','Puducherry'])]
```

```
rdf = onlyStates.drop(['Subgroup'],axis = 1)
rdf.set_index(['Area_Name', 'Year'], inplace=True)
```

```
#['Uttar Pradesh','Bihar','Delhi','Maharashtra','Tamil Nadu','Jharkhand']
rUttarPradesh = rdf.loc['Uttar Pradesh']
rUttarPradesh.head()
```

```
Year = []
rUttarPradesh = rUttarPradesh.groupby(rUttarPradesh.index).sum()
Year = rUttarPradesh.index.values.tolist()
```
